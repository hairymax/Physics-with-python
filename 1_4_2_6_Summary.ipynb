## Итоги

### Что нужно знать, чтобы начать?
* Знать, как создавать массивы: `array`, `arange`, `ones`, `zeros`.

* Узнать форму массива с помощью `array.shape`, после чего использовать слайсинг, чтобы получить различные виды массива: `array[::2]`, и т. д. Регулировать форму массива, используя `reshpe` или сглаживать его, используя `ravel`.

* Пооучить подмножество элементов массива и/или модифицировать их значения с масками

`a[a < 0] = 0`

* Знать различные операции над массивами, такие как нахождение среднего или максимального ('array.max()', 'array.mean()'). Нет нужно запоминать всё, но нужно иметь рефлекс искать в документации (онлайн документы, 'help()', 'lookfor()'!!

* Для продвинутого использования: освоить индексацию массивов целых чиссел, а также приведение. Знать больше о функциях NumPy, чтобы обходиться с различными операциями над массивами
